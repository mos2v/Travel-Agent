```python
import pandas as pd
from pathlib import Path
```

```python
base_path = Path.cwd()
```

```python
df1 = pd.read_csv(base_path / '../data/places_details.csv')
df2 = pd.read_csv(base_path / '../data/places_details1.csv')
df3 = pd.read_csv(base_path / '../data/places_details2.csv')
```

```python
merged_df = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
merged_df
```

```python
merged_df.duplicated(subset='id').sum()
```

```python

merged_df = merged_df.drop_duplicates(subset='id')

merged_df.to_csv(base_path / '../data/places_details_v1.csv', index=False)
```
